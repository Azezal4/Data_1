```python
arr1 = np.arange(1,10).reshape(3,3)
print(arr1)
```

```python
arr1.mean()
```

```python
print(arr1.mean(axis = 1))
print(arr1.mean(axis = 0)) #here the axis 0 is col and 1 is row 
```

```python
np.median(arr1)
```

```python
print(np.median(arr1,axis = 0)) #calculate the median of the col 
print(np.median(arr1,axis = 1)) #calculate the median of the row 
```

```python
print(arr1.reshape(-1,3)) #slicing

print(arr1.reshape(-1,1)) #this concept is mostly used in planes 
```
