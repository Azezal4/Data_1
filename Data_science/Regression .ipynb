# Linear Regression

```python
import numpy as np 
from scipy.stats import linregress
import matplotlib.pyplot as plt 
```

```python
x = np.random.normal(5,1,100)
y = x**2
```

```python
plt.scatter(x,y,s = 10)
plt.show()
```

```python
slope, intercept, r_value, p_value, std_error = linregress(x,y)
```

```python
def predict(x):
    return slope*x+intercept 
```

```python
a = np.arange(0,8,0.01)
plt.scatter(x,y,s=5)
plt.plot(a,predict(a),color='red')
plt.show()
```

# Polynomial Regression 

```python
eq = np.poly1d(np.polyfit(x,y,2))
```

```python
plt.scatter(x,y)
plt.plot(a,eq(a),c = 'r')
plt.show()
```

```python
y1 = np.random.normal(20,2,100)
eq1 = np.poly1d(np.polyfit(x,y1,8))
plt.scatter(x,y1)
plt.plot(a,eq1(a),c = 'r')
plt.show()
```
