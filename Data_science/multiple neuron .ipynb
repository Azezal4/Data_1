```python
import numpy as np
import pandas as pd 
from keras.models import Sequential
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from keras.layers import Dense, Dropout 
from sklearn.preprocessing import StandardScaler
```

```python
data = pd.read_csv('dataset_bank.csv')
data.head()
```

```python
data = data.drop(['RowNumber','CustomerId','Surname'],axis = 1)
```

```python
categorical_Data = ['Geography','Gender','HasCrCard','IsActiveMember']
numerical_Data = ['CreditScore','Tenure','Age','Balance','EstimateSalary']
```

```python
for i in categorical_Data:
    data[i] = pd.Categorical(data[i])
data = pd.get_dummies(data)
data.head()
```

```python
a = data.columns[:6]
```

```python
scl = StandardScaler()
for i in range(6):
    data[a[i]] = scl.fit_transform(data.iloc[:,i:i+1])
data.head()
```

```python
data['Exited']
```

```python
X = data.drop('Exited', axis = 1)
y= data['Exited']
```

```python
X.head()
```

```python
X.shape
```

```python
X_train , X_test, y_train, y_test = train_test_split(X,y,test_size = 0.3)
```

```python
X_train.shape
```

```python
nn = Sequential()
```

```python
nn.add(Dense(units=15,activation='relu',input_dim = 15, kernel_initializer='uniform'))

nn.add(Dense(units=9, activation = 'relu',kernel_initializer='uniform'))
nn.add(Dropout(rate=0.3))

# nn.add(Dense(units = 8, activation = 'relu',kernel_initializer='uniform'))
# nn.add(Dropout(rate=0.3))

#nn.add(Dense(units=8, activation = 'relu',kernel_initializer='uniform'))
# nn.add(Dropout(rate=0.3))

#nn.add(Dense(units=8, activation = 'relu',kernel_initializer='uniform'))

nn.add(Dense(units=1, activation = 'sigmoid',kernel_initializer='uniform'))

nn.compile(optimizer = 'adam',loss = 'binary_crossentropy',metrics = ['accuracy'])
```

```python
nn.fit(X_train,y_train,epochs =10, batch_size = 2000, validation_split = 0.01 ) #epoches does rerun 
```

```python
nn.evaluate(X_test,y_test)
```

```python
y_pred = nn.predict(X_test)
y_pred
```
