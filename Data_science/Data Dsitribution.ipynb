# Data Distribution 


Unifrom Distribution 

```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
uniform_data = np.random.uniform(1000,10000,50000) #taking uniform data 
plt.hist(uniform_data,50)
plt.show()
```

Normal Distribution

```python
normal_data = np.random.normal(5000,1000,50000) #normally distributed data 
plt.hist(normal_data,50)
plt.show()
```
